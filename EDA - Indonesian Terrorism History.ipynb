```python
fp = 'C:/Intern/idn_adm_bps_20200401_shp/idn_admbnda_adm1_bps_20200401.shp'
map_df = geopandas.read_file(fp)
```

```python
# Select relevant columns
data = df_clean[['eventid','latitude', 'longitude','date','attacktype1_txt','targtype1_txt','weaptype1_txt','nkill']]
```

```python
gdf = geopandas.GeoDataFrame(
    data, geometry=geopandas.points_from_xy(data.longitude, data.latitude), crs="EPSG:4326"
)
```

```python
gdf['PROVINCE'] = ''
for idx in range(map_df.shape[0]):
    # For every address, find if they reside within a province
    pip = gdf.within(map_df.loc[idx, 'geometry'])
    if pip.sum() > 0: 
        # If some of the addresses reside at map_df.loc[idx]
        province = map_df.loc[idx, 'ADM1_EN']
        gdf.loc[pip, 'PROVINCE'] = province

# Drop rows where 'PROVINCE' is empty
gdf = gdf[gdf['PROVINCE'] != '']
```

```python
def scoring_attack_type(score_attack):
    if score_attack in ['Hostage Taking (Barricade Incident)', 'Unarmed Assault','Hostage Taking (Kidnapping)', 'Unknown']:
        return int(1)
    if score_attack in ['Facility/Infrastructure Attack', 'Hijacking']:
        return int(2)
    if score_attack in ['Armed Assault', 'Bombing/Explosion', 'Assassination']:
        return int(3)

def scoring_targ_type(score_targ):
    if score_targ in ['Religious Figures/Institutions', 'Educational Institution', 'NGO', 'Business', 'Journalists & Media', 'Terrorists/Non-State Militia', 'Violent Political Party', 'Unknown']:
        return int(1)
    if score_targ in ['Airports & Aircraft', 'Transportation', 'Utilities', 'Maritime', 'Private Citizens & Property', 'Tourists', 'Food or Water Supply']:
        return int(2)
    if score_targ in ['Government (General)', 'Military', 'Government (Diplomatic)', 'Police']:
        return int(3)

def scoring_weap_type(score_weap):
    if score_weap in ['Melee', 'Unknown', 'Other']:
        return int(1)
    if score_weap in ['Incendiary', 'Chemical']:
        return int(2)
    if score_weap in ['Firearms', 'Explosives']:
        return int(3)

def scoring_nkill(score_nkill):
    if score_nkill in range(0, 6):
        return int(1)
    if score_nkill in range(5, 20):
        return int(2)
    if score_nkill in range(20,110):
        return int(3)

def alert_level(alert_score):
    if alert_score in range(3, 7):
        return 'Low Level'
    if alert_score in range(6, 10):
        return 'Medium Level'
    if alert_score in range(9, 13):
        return 'High Level'
```

```python
gdf['Attack Score'] = gdf['attacktype1_txt'].apply(scoring_attack_type)
gdf['Target Score'] = gdf['targtype1_txt'].apply(scoring_targ_type)
gdf['Weap Score'] = gdf['weaptype1_txt'].apply(scoring_weap_type)
gdf['Kill Score'] = gdf['nkill'].apply(scoring_nkill)
gdf['Threat Score'] = gdf['Attack Score'] + gdf['Target Score'] + gdf['Weap Score'] + gdf['Kill Score']
gdf['Alert Level'] = gdf['Threat Score'].apply(alert_level)
```

**Exploratory Data Analysis**

```python
# Counting the number of incidents per province
location_incident_count = gdf['PROVINCE'].value_counts()

# Plotting
plt.figure(figsize=(10, 6))
ax = location_incident_count.head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Province by Number of Incidents')
plt.xlabel('Province')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=45)
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
plt.tight_layout()
plt.show()
```

```python
# Calculate success rate for each location
success_rate = gdf.groupby('PROVINCE')['success'].mean()

# Plotting success rate
plt.figure(figsize=(10, 6))
ax2 = success_rate.loc[location_incident_count.head(10).index].plot(kind='bar', color='green')
plt.title('Success Rate in Top 10 Locations')
plt.xlabel('Location')
plt.ylabel('Success Rate')
plt.xticks(rotation=45)
for p in ax2.patches:
    ax2.annotate(f'{p.get_height():.2f}', (p.get_x() * 1.005, p.get_height() * 1.005))

plt.tight_layout()
plt.show()
```

```python
# Examining the trend of incidents over the years
gdf['year'] = pd.to_datetime(gdf['date']).dt.year
yearly_incident_count = gdf['year'].value_counts().sort_index()

# Plotting
plt.figure(figsize=(10, 6))
ax = yearly_incident_count.plot(kind='line', marker='o', linestyle='-', color='red')
plt.title('Number of Incidents Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Incidents')
for i, label in enumerate(list(yearly_incident_count.index)):
    score = yearly_incident_count[label]
    plt.annotate(score, (label, score), textcoords="offset points", xytext=(0,10), ha='center')
plt.tight_layout()
plt.show()
```

```python
# Calculate success rate for each location
success_rate = gdf.groupby('year')['success'].mean()

# Plotting success rate
plt.figure(figsize=(10, 6))
ax2 = success_rate.loc[yearly_incident_count.index].plot(kind='line', marker='o', linestyle='-', color='red')
plt.title('Success Rate of Incidents Over the Years')
plt.xlabel('Year')
plt.ylabel('Success Rate')
plt.xticks(rotation=45)
for p in ax2.patches:
    ax2.annotate(f'{p.get_height():.2f}', (p.get_x() * 1.005, p.get_height() * 1.005))

plt.tight_layout()
plt.show()
```
